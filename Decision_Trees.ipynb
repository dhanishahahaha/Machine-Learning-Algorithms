# **Decision Trees**

A decision tree is a non-parametric supervised learning algorithm, which is utilized for both classification and regression tasks. It has a hierarchical, tree structure, which consists of a root node, branches, internal nodes and leaf nodes.

**Logic behind Decision Trees in 4 easy steps:**

* Begin with your training dataset , which should have some feature variables and classification or regression output.

* Determine the *best feature* in the dataset to split the data on.

* Split the data into subsets that contain the correct values for this best feature. This splitting basically defines a node on the tree i.e each node is a splitting point based on a certain feature from our data.

* Recursively generate new tree nodes by using the subset of data created from step 3.

# Entropy:

**What is Entropy** in Decision Tress?

In most Layman terms , Entropy is the measure of disorder. You can also call it measure of purity/impurity.


image.png

We can also say, '*More knowledge, Less entropy*'. This means that if there is more randomness then it becomes more difficult to predict.

image.png

Using this formula, we can calculate the entropy of different datasets to measure the randomness.

We can also add the number of attributes in the formula.

**Observations:**

* More the uncertainity, more the entropy.

* For a 2 class problem, the min entropy is 0 and the max entropy is 1.

* For more than 2 classes, the min entropy is 0 but the max entropy can be greater than 1.

* Both log2 and loge can be used to calculate the entropy.



When we talk about entropy for continous variables, how would we know which variable have higher entropy?

 --> By looking at the peaks of the variable distribution. *Whichever is less peaked will have a higher entropy.*

# Information Gain:

Information Gain is a metric used to train the Decision Trees. Specifically, this metric measures the quality of the split.

Information Gain is based on the decrease in the entropy after the dataset is split on an attribute.Constructing a decision tree is all about finding attributes that returns the highest information gain.

**Information Gain formula:**

IG = E(parent) - {Weighted Average} * E(Children)

**Steps to calculate Information Gain:**

* Calculate the entropy of the entire dataset.
* Divide the data into child branches as per same attribute values and calculate the entropy for every branch.
* Calculate the Weighted entropy of children.
* Calculate Information Gain using the formula.
* Calculate the Information Gain for all columns, whichever column has the higest information gain the algorithm will select that column to split the data.

* Find Information Gain recursively, Decision Tree then applies a Greedy algorithm to find Information Gain at every level of the tree.
weighted entropy = no. of rows in 1 child branch/original number of rows + same for child 2 + child 3 +..+n

**note**: if the entropy of any branch is zero, it is considered as a leaf node.

# How to handle numerical data with Decision Trees?

Suppose we had a dataset that had 2 columns, **User Rating and Downloaded**. 

The User Rating contains Numerical Data and the Downloaded columns has values as Yes or No.

Since the numerical data is unique and different for different users, it is not possible to categorise them and make branches...it would create tooo many branches!!

Then what could be the possible solution if a column had unique values?

# solution:

**Step 1: Sort the data on the basis of Numerical column.**

**Step 2: Split the entire data on the basis of every value of user_rating**

for example if the values in user ratings were: 1.9,2.4,3.3,5.0...
so we will do it as 
 * if rating > 1.9 (this would split the dataset into 2 parts, 1 row in dataset 1 and rest of the rows in dataset 2)
 * if rating > 2.9 ( again this would split the dataset into 2 parts, 2 rows in dataset 1 and rest of the rows in dataset 2)
 * if rating > 3.3 (again, 3 rows in dataset 1 and rest of the rows in dataset 2)
 * if rating > 5.0 (4 rows in dataset 1 and rest in dataset 2)

we split the data into 2 parts n times if there were n values for user ratings.

**Step 3: Calculate the entropy E1 and E2 for the 2 datasets that were created in each case**.

**Step 4: Calculate the weighted entropy of E1 and E2 as WE for all cases as WE1 , WE2 , WE3...WEn.**

**Step 5: Now calculate  the information gain for every case as IG as IG1 , IG2 , IG3.....IGn.**

**Step 6: Now from all the information gain values, see the maximum value of IG.
suppose the max value of IG we got in IG3, so we will set it as split criteria data.**

**Step 7: Do it recursively until we get all the leaf nodes.**


We do all these steps during the training phase not while testing/deploying.
Thus, the Train time complexiety for this algorithm is higher but the Test time complexiety is of order of log(n).

Due to this complex sorting for Numerical Data in Decision Trees, it is considered more apt for Categorical Data. So, if the data has too many Numerical columns, we must avoid Decision Trees on those type of datasets.

#split/depth Hyper Parameter in Decision Tree Classifier:

Depth of the tree:

(Overfitting - performs well on train data but poorly on test data.)

max_depth = None 

means the algorithm will work until it extracts every leaf node in the tree but then there is a huge chance of overfitting.


(Underfitting - performs poorly on both test and train data)

max_depth = 1

this means we are only going to split once. But this means all the data will be categorised on the basis of one split only.






* Thus, too much value or None value of max_depth leads to Overfitting and very less value of max_depth leads to Underfitting.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from google.colab import files
```

```python
file = files.upload()
```

```python
def analyzer(max_depth):
  data = pd.read_csv('social network ads.csv')
  x = data.iloc[:,0:2].values
  y = data.iloc[:,-1].values


  clf = DecisionTreeClassifier(max_depth=max_depth)
  clf.fit(x,y)

  a = np.arange(start = x[:,0].min()-1 , stop = x[:,0].max()+1 , step = 0.1)
  b = np.arange(start = x[:,1].min()-1 , stop = x[:,1].max()+1 , step = 100)

  xx,yy = np.meshgrid(a ,b)

  input_array = np.array([xx.ravel(), yy.ravel()]).T

  labels = clf.predict(input_array)

  plt.contourf(xx , yy , labels.reshape(xx.shape) , alpha=0.5)
  plt.scatter(x[:,0] , x[:,1] , c = y)
```

```python
analyzer(max_depth=None)        #overfitting
```

```python
analyzer(max_depth = 1)         #underfitting
```

```python
analyzer(max_depth = 2)
```

```python
analyzer(max_depth = 3)
```

```python
analyzer(max_depth = 4)      #now this seems an ideal value for max_depth because it is able to label the points correctly even outside the split.
```

## Code

```python
from pandas_datareader import data
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
df = pd.DataFrame(wine.data)
```

```python
df.columns = wine.feature_names
df['proline'] = wine.target
```

```python
df.head()
```

```python
df.shape
```

```python
x = df.iloc[:,0:12]
y = df.iloc[:,12]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rt = DecisionTreeRegressor(criterion = 'squared_error',max_depth=5)
```

```python
rt.fit(x_train , y_train)
```

```python
y_pred = rt.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test , y_pred)
```

Hyperparameter tuning: we can either use Grid Search CV or Random Search CV

```python
param_grid = {
    'max_depth' : [2,4,8,10,None],
    'criterion' : ['average_error' , 'squared_error'],
    'max_features' : [0.25,0.5,1.0],
    'min_samples_split' : [0.25,0.5,1.0]
} 
```

```python
reg = GridSearchCV(DecisionTreeRegressor(),param_grid=param_grid)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.best_score_
```

```python
reg.best_params_
```

Feature Importance: helps for feature selection and gives the list of most important to unimportant columns from the data.

```python
for importance, name in sorted(zip(rt.feature_importances_, x_train.columns),reverse=True):
  print(name , importance)
```
